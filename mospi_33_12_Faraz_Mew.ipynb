```
import pandas as pd
import numpy as np
df1 = pd.read_csv("/content/drive/My Drive/Copy of mospi_33_12.csv")
df1.head(5)
def river_info(df1,name):
  df1 = df1[df1["river_names"] == name]
  index = df1.parameters.unique()
  df1 = df1.reset_index(drop =True)
  df1.reset_index(inplace =True)
  new_df = pd.DataFrame()
  ind = 0
  for i in range(2001,2012):
    l1 = np.array(())
    for j in range(len(index)):
      l1 = np.append(l1,df1["value"][ind])
      ind = ind+1
    new_df[str(i)] = l1
  new_df=new_df.set_index(index)
  return new_df
print(df1.river_names.unique())
river = input("Enter river name: ").capitalize()
river_df = river_info(df1,river)
river_df.head(10)
```

```
def isinteger(a):
    try:
        int(a)
        return True
    except ValueError:
        return False
ch = True
while ch == True:
  print("Please enter 1 to save in .xlsx format 0 to exit")
  choice = input("Please enter your choice  ")
  value = isinteger(choice)
  if value == True:
    if int(choice) == 1:
      river_df.to_excel("mospi_33_12_Faraz_Mew.xlsx")
      print("DataFrame is saved in xlsx format")
      break
    elif int(choice) == 0:
      break
    else:
      print("Wrong choice please try again")
      ch = True
  else:
    print("Wrong choice please try again")
  ch = True
```
